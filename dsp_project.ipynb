### Loading the datasets into pandas DataFrames
* Using 'pandas', 'glob', 'matplotlib' libraries and 'list'.

##### The following codes include:

* A check to ensure CSV files are found.
* A loop to read each CSV file into a DataFrame, with error handling.
* A check to ensure that there are at least 5 DataFrames before attempting to assign them to variables.
* If the csv_files list is empty or contains fewer than 5 files, the appropriate message will be printed, and the code will not attempt to access non-existent elements.

```python
# importing required libraries
import pandas as pd                     
import glob                             
import matplotlib.pyplot as plt 
import numpy as np        

# Define the path to the CSV files
csv_files_path = 'archive/*.csv'


# Use glob to get all the CSV file paths    
csv_files = glob.glob(csv_files_path)

# Create a list to hold the DataFrames

if not csv_files:
    print("No CSV files found in the specified directory.")
else:
    dataframes = []
    for file in csv_files:
        try:
            df = pd.read_csv(file)
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")

    print(f'Number of CSV files: {len(dataframes)}')

    # Check if we have at least 5 dataframes
    if len(dataframes) < 5:
        print("Not enough dataframes. Expected at least 5, but found:", len(dataframes))
    else:
        # Assign the DataFrames to variables by index
        df2015 = dataframes[0]
        df2016 = dataframes[1]
        df2017 = dataframes[2]
        df2018 = dataframes[3]
        df2019 = dataframes[4]
        print("DataFrames assigned successfully.")
```

### Impact of other factors (GDP, Health, etc.) on Happiness Score

```python
# checking the impact of other factors on Happiness Score using the 5 happiest countries in 2015
impact2015 = df2015_new_sorted.head(5)
impact2015
```

```python
# bar plots of impacts of other factors in 2015 
n = len(impact2015)
n
# Position of bars on the x-axis
ind = np.arange(n)

# Width of a bar
width = 0.2
# Create a bar plot
plt.bar(ind, impact2015['H_Score'], width, label='HS')
plt.bar(ind + width, impact2015['GDP'], width, label='GDP')
plt.bar(ind + 2*width, impact2015['Health'], width, label='Health')
plt.bar(ind + 3*width, impact2015['Trust'], width, label='Trust')

# Add title and labels
plt.title('Impact of other factors on H_Score for happiest countries in 2015')
plt.xlabel('Country')
plt.ylabel('Values')

# Add xticks on the middle of the bars
plt.xticks(ind + width, impact2015['Country'])

# Add a legend
plt.legend()

# Show the plot
plt.show()  
```

```python
# creat a new variable 'region_happiness' to hold the average happiness score by region
region_happiness = df2015_new_sorted.groupby(['Region'])['H_Score'].mean()
region_happiness
```

```python
# convert 'region_happiness' to frame and reset the index
region_happiness = region_happiness.to_frame('Avg_Happiness').reset_index()  # type: ignore
region_happiness
```

```python
# bar plot of average happiness by region
region_happiness.plot(kind='bar', x='Region', y='Avg_Happiness', color='purple')

plt.title('Average Happiness Scores across Regions')
plt.xlabel('REGION', color='blue')
plt.ylabel('AVERAGE HAPPINESS SCORE', color='green')

plt.show()  # display the plot
```
